# Functions & Variables



A function performs a specific set of pre-defined instructions, there are many that come with the base Python language. Print is one of the most common functions in python - taking whatever message is passed into it, and outputting it to your screen.

````python
# This is a comment - if you start a line with the # symbol it means that the whole line will not be run, and will just show as green text.
# They're a nice way to leave notes for us to come back to, or help other people understand what we were thinking
print("Hello Geography Club!")

# If I want to stop a line of code from running, I can "comment it out" and put a # in front of the line. In the line below, the Hello World message won't run
# print("Hello World")
````

In general, we can think of functions as following this form: `function(argument)`. The actual task you're doing is described in name by the `function` and whatever the `function` needs as an input is placed in the parentheses as the `argument`.


A variable is a container for storing a data value. Let's check out a couple:




````python
# This is an example of a string (used for raw text)
message = "Hello Geography Club!"
print(message)

# Here's an integer (used for whole numbers)
favorite_number = 7
print(favorite_number)

# A float (for floating-point decimal)
pi = 3.14159265

# A Boolean (true or false)
UCSB_Student = True
print(UCSB_Student)

# some expressions return a boolean

print(3 > 5)

# What if I want to change my variable...

message = "Goodbye Geography Club!"
print(message)

# but variables are case sensitive, so if I re-assign a Message variable then nothing happens

Message = "We're so back"
print(message)

# Python also works as a great calculator

a = 7
b = 5
print(a + b)
print(a - b)
print(a * b)
print(a / b)
````

````python
# We can also store several elements in a list, indicated with brackets and has items separated with commas
groceries = ["bananas", "grapes", "fruits"]
print(groceries)

# They don't all have to be the same data type - we can keep all sorts of stuff in the same listed
random_stuff = [favorite_number, "spruce", UCSB_Student]
print(random_stuff)
````

Python (and R) are very unique in that you don't have to explicitly say what variable you want to be using - it'll automatically pick up on variable type.

We'll start by loading in some geographic information using geopandas, which is going to be your bread and butter. This builds upon the dataframes that pandas uses - but supports the use of the geographic objects. This means we can now store points, lines, polygons, and more - all with the standard methods of manipulation that are supported in GIS software.

````python
!pip install geopandas
````

````python
import geopandas as gpd

# Data pulled from here: https://www.naturalearthdata.com/downloads/110m-cultural-vectors/

# here's where our data is located
states_path = "Data/US-States/ne_110m_admin_1_states_provinces.shp"

states = gpd.read_file(states_path)
states.head()
````

````python
# Let's give it a quick look
states.plot()
````

````python
# Let's also see what coordinate reference system (CRS) we're working with
print(states.crs)
````

````python
# That's not great, we should have a defined CRS. We can take a look at the Natural Earth Website to see what CRS their data is stored in - it's in WGS 84
# Lets manually set the CRS now
states = states.set_crs("EPSG:4326")
print(states.crs)
states.plot()
````

While I know they're lovely, let's drop Alaska and Hawaii to focus on the Continental United States.

````python
# This is how we can select an individual row using a desired attribute value, using the == operator to only select a row where the name is equal to California
# while = might seem more intuitive, we save that for assigning things not comparisons
california = states[states["name"] == "California"]
california.plot()
````

````python
# It's now time to remove Alaska and Hawaii
# Here, the ~ means not, so we're now going to exclude whatever selection we make in the brackets
states_conus = states[~states['name'].isin(['Alaska', 'Hawaii'])]
states_conus.plot()
````

````python
# we're going to extract the axes of the plot to do a little customization
ax = states_conus.plot()

# We're going to add a title to the axes object we've made and take a look
ax.set_title('Continental United States')
ax.plot()
````

Packages, like geopandas, are often dependent on other associated packages to run - we call these dependencies. We don't need to know too much about that, but sometimes the associated packages give you a bit more control and functionality, like matplotlib, which is the library that let us make that plot above.

````python
# We use matplotlib.pyplot instead of all of matplotlib because we're only interested in a small collection of plotting functions out of everything the package provides.
import matplotlib.pyplot as plt

# Create a simple list of numbers
data_points = [1, 3, 2, 5, 4]

# Generate a line graph
plt.plot(data_points)

# Add title and labels
plt.title("Simple Line Plot")
plt.xlabel("Index")
plt.ylabel("Value")

# Display the plot
plt.show()
````

````python
#fig — the entire figure/window, controls overall size and saving
# ax — the individual plot area where data gets drawn, controls axes, titles, labels
fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax = ax, color = "lightgreen", alpha = 1, edgecolor = "black", linewidth = 0.5)
ax.set_title("Continental United States")
plt.tight_layout()
plt.show()
````

````python
# We can build upon this in a bunch of different ways, like adding labels -- these need some work
fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax = ax, color = "lightgreen", edgecolor = "black", linewidth = 0.5)
states_conus.apply(lambda x: ax.annotate(x['name'], xy=x.geometry.centroid.coords[0], ha='center', fontsize=6), axis=1)
plt.tight_layout()
plt.show()
````

Let's bring in the UFO data now.

````python
ufo_df = pd.read_excel("Data/UFOs_coord.xlsx")
ufo_df.head()
````

Unlike the shapefile we loaded earlier, this data comes to us as a plain spreadsheet - but it does have `lat` and `lng` columns that we can use. We can use `gpd.points_from_xy()` to convert those coordinates into geometry objects, turning our regular dataframe into a GeoDataFrame.

````python
ufo_gdf = gpd.GeoDataFrame(ufo_df, geometry=gpd.points_from_xy(ufo_df['lng'], ufo_df['lat']), crs="EPSG:4326")

# trim to the continental US using what we know about US coordinates
ufo_conus = ufo_gdf[(ufo_gdf['lat'] >= 24) & (ufo_gdf['lat'] <= 50) & (ufo_gdf['lng'] >= -125) & (ufo_gdf['lng'] <= -66)]

print(f"{len(ufo_conus)} UFO sightings in the continental US")
ufo_conus.head()
````

````python
fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.5)
ufo_conus.plot(ax=ax, color="dodgerblue", markersize=1.5, alpha=0.4)
ax.set_title("UFO Sightings in the Continental United States")
plt.tight_layout()
plt.show()
````

That's not so great - we've still got points that don't fall within state boundaries. Lets improve upon the coordinate filters we used earlier with a spatial join.

````python
ufo_conus = gpd.sjoin(ufo_gdf, states_conus[['geometry']], how='inner', predicate='within')
````

````python
fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.5)
ufo_conus.plot(ax=ax, color="dodgerblue", markersize=1.5, alpha=0.4)
ax.set_title("UFO Sightings in the Continental United States (Spatial Join)")
plt.tight_layout()
plt.show()
````

UFOs apparently come in all shapes and sizes - let's see what's out there and try filtering to just one.

````python
ufo_conus['Shape'].value_counts()
````

````python
# Swap out 'Sphere' for anything from the list above!
ufo_filtered = ufo_conus[ufo_conus['Shape'] == 'Sphere']

fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.5)
ufo_filtered.plot(ax=ax, color="dodgerblue", markersize=3, alpha=0.6)
ax.set_title("Sphere-Shaped UFO Sightings")
plt.tight_layout()
plt.show()
````

````python
# Plotting Top 10 Shapes

top_shapes = ufo_conus['Shape'].value_counts().nlargest(10).index
ufo_top = ufo_conus[ufo_conus['Shape'].isin(top_shapes)]


fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", alpha = 0.5, edgecolor="black", linewidth=0.5)
ufo_top.plot(ax=ax, column='Shape', cmap='tab20', markersize=2, alpha=0.6, legend=True)
ax.set_title("UFO Sightings by Shape (Top 10)")
plt.tight_layout()
plt.show()
````

Now let's see what Bigfoot has been up to.

````python
bigfoot_df = pd.read_csv("Data/bfro_reports_geocoded.csv")
bigfoot_df.head()
````

````python
# Some reports are missing coordinates - we'll drop those
bigfoot_df = bigfoot_df.dropna(subset=['latitude', 'longitude'])

bigfoot_gdf = gpd.GeoDataFrame(bigfoot_df, geometry=gpd.points_from_xy(bigfoot_df['longitude'], bigfoot_df['latitude']), crs="EPSG:4326")

bigfoot_conus = gpd.sjoin(bigfoot_gdf, states_conus[['geometry']], how='inner', predicate='within')

print(f"{len(bigfoot_conus)} Bigfoot sightings in the continental US")
bigfoot_conus.head()
````

````python
fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.5)
bigfoot_conus.plot(ax=ax, color="saddlebrown", markersize=2, alpha=0.5)
ax.set_title("Bigfoot Sightings in the Continental United States")
plt.tight_layout()
plt.show()
````

The dataset also includes a `season` column and a `classification` - Class A means a direct sighting, Class B means something more indirect like tracks or sounds. Try filtering by season, or by class!

````python
bigfoot_conus['season'].value_counts()
````

````python
# Swap out 'Summer' for any season from the list above!
bigfoot_filtered = bigfoot_conus[bigfoot_conus['season'] == 'Summer']

fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.5)
bigfoot_filtered.plot(ax=ax, color="saddlebrown", markersize=2, alpha=0.5)
ax.set_title("Summer Bigfoot Sightings")
plt.tight_layout()
plt.show()
````

Let's take a look at them all together now - are there areas more prone to the paranormal?

````python
fig, ax = plt.subplots(figsize=(12, 8))
states_conus.plot(ax=ax, color="lightgray", edgecolor="black", linewidth=0.5)
ufo_conus.plot(ax=ax, color="dodgerblue", markersize=1.5, alpha=0.4, label="UFO Sightings")
bigfoot_conus.plot(ax=ax, color="saddlebrown", markersize=2, alpha=0.5, label="Bigfoot Sightings")
ax.legend()
ax.set_title("UFO vs. Bigfoot Sightings in the Continental United States")
plt.tight_layout()
plt.show()
````
